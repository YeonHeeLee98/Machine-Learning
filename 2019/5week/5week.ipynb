```python
import numpy as np
import pickle
import os
import pandas as pd

def makeMatrix():
    np.random.seed(20175327)
    matrix=np.round(np.random.random(40)*10)
    matrix[matrix == 0] = np.NaN
    return matrix
    
def savefile(file_path,matrix):
    f=open("matrix.pkl","wb") 
    pickle.dump(matrix,f)
    f.close()
    
def loadfile(file_path):
    try:
        f=open("matrix.pkl","rb")
        data=pickle.load(f)
    except FileNotFoundError as e:
        print(e)
    return data

def nanToZero(matrix):
    f=open("matrix.pkl","rb")
    data=pickle.load(f)
    np.nan_to_num(data,copy=False)  
    return data

def makeShape(matrix,row,col):
    make = 1
    shape = matrix.shape
    for s in shape:
        make = make*s
    if make != row*col:
        return 0
    a = np.zeros((row,col))
    for i in range(row):
        a[i] = matrix[i*col:(i+1)*col]
    return a

def matMul(mat1,mat2):
    if mat1.shape[1] != mat2.shape[0]:
        return 0
    a = np.zeros((mat1.shape[0],mat2.shape[1]))
    for i in range(mat1.shape[0]):
        for j in range(mat2.shape[1]):
            a[i][j] = np.sum(mat1[i,:] * mat2[:,j])
    return a

def makePandas(matrix):
    shape = matrix.shape
    frame = pd.DataFrame()
    index = [str((i+1)) + '행' 
             for i in range(shape[0])]
    col = [str((i+1)) + '열' 
           for i in range(shape[1])]
    for i in range(0,len(col)):
        frame[col[i]] = matrix[:,i]
    frame.index = index
    return frame

if __name__=="__main__":
    file_path=os.getcwd()+"/matrix.pkl"
    matrix=makeMatrix()
    print(matrix)
    savefile(file_path,matrix)
    
    del matrix
    matrix=loadfile(file_path)
    print(matrix)
    
    matrix=nanToZero(matrix)
    print(matrix)
    
    mat1=makeShape(matrix,8,5)
    print(mat1)
    mat2=makeShape(matrix,5,8)
    print(mat2)
    
    result=matMul(mat1,mat2)
    print(result)

    frame=makePandas(result)
    display(frame)
```
